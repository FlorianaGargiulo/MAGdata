```python
redDf=thisDf[thisDf['feature']!=' http://www.w3.org/1999/02/22-rdf-syntax-ns#type']
redDf=redDf[redDf['feature']!=' rank']
redDf=redDf[redDf['feature']!=' estimatedCitationCount']
```

```python
feat=list(set(redDf['feature']))
```

```python
feat
```

```python
dictFeat={}
for f in feat:
    #dictFeat[f]={}
    df=redDf[redDf['feature']==f]
    dictFeat[f]=dict(zip(df['id'],df['value']))
```

```python
dictFeat[' title']
```

```python
ids=list(set(redDf['id']))
```

```python
len(ids)
```

```python
dictPapers=[]
for p in ids:
    thisdict={}
    thisdict['id']=p
    for f in dictFeat.keys():
        try:
            thisdict[f]=dictFeat[f][p]
        except:
            continue
    dictPapers.append(thisdict)
        
     
```
